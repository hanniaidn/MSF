```python
import numpy as np
import matplotlib.pyplot as plt 
import control as ctrl
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import tkinter as tk
import get_kr
```

```python
## parametros iniciales 
m=1 # kg
b=10 # N s/m
k=20 # N/m

#systema 
A=np.array([[0,1],
           [-k/m,-b/m]])
B=np.array([[0],
            [1/m]])
C=np.array([1,0])
D=0

#ss del system 
ss_sys=ctrl.ss(A,B,C,D)

#tf del systema 
tf_sys=ctrl.ss2tf(ss_sys)

#mostar la tf del systema 
tf_sys
```

```python
#cambio en kp

kp=575

sys=ctrl.feedback(tf_sys*kp)
sys

t, y=ctrl.step_response(sys)

plt.figure()
plt.plot(t,y)
    
```

```python
# cambio en kd
kd=5
kp=10

sys=ctrl.feedback(tf_sys*kp*kd)
sys

t, y=ctrl.step_response(sys)

plt.figure()
plt.plot(t,y)

    
```
